# Suponha agora que uma amostra de 50 embalagens foi selecionada e que a média amostral observada foi de 503,24 ml. Esse valor de média amostral é suficientemente maior que 500 ml para nos fazer rejeitar a hipótese de que a média do processo é de 500 ml ao nível de significância de 5%?**

```r
amostra <- c(509, 505, 495, 510, 496, 509, 497, 502, 503, 505, 
           501, 505, 510, 505, 504, 497, 506, 506, 508, 505, 
           497, 504, 500, 498, 506, 496, 508, 497, 503, 501, 
           503, 506, 499, 498, 509, 507, 503, 499, 509, 495, 
           502, 505, 504, 509, 508, 501, 505, 497, 508, 507)
```

```r
significancia <- 0.05
```

```r
confianca <- 1 - significancia
```

```r
n <- nrow(amostra)

n
```

**Passo 1 - formulação das hipóteses**

*H0 : média = 500*

*H1 : média != 500*

**Passo 2 - escolha da distruição amostral adequada**

*N > 30. Sigma desconhecido. Então utilizo distribuição normal junto com o desvio padrao da amostra*

**Passo 3 - fixação da significância do teste (alfa)**

```r
probabilidade <- 0.5 + confianca / 2 

probabilidade
```

```r
z_alpha_2 <- qnorm(probabilidade)

z_alpha_2
```

```r
z >= z_alpha_2
```

```r
if(z <= - z_alpha_2 || z >= z_alpha_2){
    
    "Rejeitar H0"
    
} else {
    
    "Aceitar H0"
    
}
```

```r
p_valor <- 2 * (1 - pnorm(z))

p_valor
```

```r
t <- (82 - media) / (sd_amostra / sqrt(n))

t
```

```r
calcular_p_valor <- function(media_amostra, media, sd_amostra, n){
    
    t <- (media_amostra - media) / (sd_amostra / sqrt(n))
    
    p_valor <- pt(t, n - 1, lower.tail = T)
    
    return(p_valor)
    
}
```

```r
for(i in c(82, 70)){
    
    p_valor <- calcular_p_valor(i, media, sd_amostra, n)
    
    print(sprintf("Para a média amostral de %d, o valor de p é %.2f", i, p_valor))
    
    if(p_valor <= significancia){
        
        print("Rejeitar H0")
        
    } else{
        
        print("Aceitar H0")
        
    }
    
}
```

# Antes de cada partida do campeonato nacional de futebol, as moedas utilizadas pelos árbitros devem ser verificadas para se ter certeza de que não são viciadas, ou seja, que não tendam para determinado resultado.        Cara | Coroa
Observado 17 | 33

Esperado 25  | 25

# A um nível de significância de 5%, é possível afirmar que a moeda não é honesta, isto é, que a moeda apresenta uma probabilidade maior de cair com a face CARA voltada para cima?

```r
p <- c(0.005, 0.01, 0.025, 0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.975, 0.95, 0.99, 0.995)
df <- seq(1, 30, by = 1)
probabilidade <- c()
for(i in df){
    for(j in p){
        probabilidade <- c(probabilidade, round(qchisq(j, i, lower.tail = T), 4))
    }
}
tabela_chi_2 <- matrix(probabilidade, ncol=13, byrow=TRUE)
colnames(tabela_chi_2) <- format(p)
rownames(tabela_chi_2) <- format(df)
tabela_chi_2
```

```r
F_Observada <- c(17, 33)
```

```r
F_Esperada <- c(25, 25)
```

```r
significancia <- 0.05
```

```r
confianca <- 1 - significancia
```

```r
k <- 2 #Numero de eventos possiveis. binomial
```

```r
graus_de_liberdade <- k - 1
```

**Passo 1 - formulação das hipóteses**

*H0: Fcara = Fcoroa*

*H1: Fcara != Fcoroa*

**Passo 2 - fixação da significância do teste (alfa)**

```r
tabela_chi_2[1:3, ]
```

```r
chi_2_alpha <- qchisq(confianca, graus_de_liberdade)

chi_2_alpha
```

**Passo 3 - cálculo da estatística-teste e verificação desse valor com as áreas de aceitação e rejeição do teste**

```r
chi_2 <- ((F_Observada[1] - F_Esperada[1]) ** 2 / F_Esperada[1]) + ((F_Observada[2] - F_Esperada[2]) ** 2 / F_Esperada[2])

chi_2
```

**Passo 4 - Aceitação ou rejeição da hipótese nula**

```r
if(chi_2_alpha < chi_2){
    
    "Rejeitar H0"
    
} else {
    
    "Aceitar H0"
    
}
```

```r
chi_2 = 0

for(i in 1:k){
    
    chi_2 <- chi_2 + ((F_Observada[i] - F_Esperada[i]) ** 2 / F_Esperada[i])
    
}

chi_2
```

```r
if(chi_2_alpha < chi_2){
    
    "Rejeitar H0"
    
} else {
    
    "Aceitar H0"
    
}
```

**Obtendo Zteste**

```r
Z <- (T - mu_T) / sigma_T

Z
```

```r
if (Z <= - z_alpha || Z >= z_alpha){
    
    "Rejeitar H0"
    
} else {
    
    "Aceitar H0"
    
}
```

*H0: u_rj >= u_sp*

*H1: u_rj < u_sp*

*ou*

*H0: u_rj - u_sp >= 0*

*H1: u_rj - u_sp < 0*
